```python
import json
import os

def explore_json_structure(json_file_path, max_depth=5, sample_limit=2):
    """
    Recursively explore the structure of a JSON file, showing keys at different nesting levels
    
    Parameters:
    - json_file_path: Path to the JSON file
    - max_depth: Maximum depth to explore
    - sample_limit: Number of items to show at each level
    """
    print(f"Exploring JSON structure: {os.path.basename(json_file_path)}")
    
    # Load the JSON file
    try:
        with open(json_file_path, 'r') as file:
            data = json.load(file)
        
        # Start recursive exploration
        explore_structure(data, max_depth=max_depth, sample_limit=sample_limit)
            
    except json.JSONDecodeError as e:
        print(f"Error parsing JSON: {e}")
    except Exception as e:
        print(f"Error: {e}")

def explore_structure(data, prefix="", depth=0, max_depth=5, sample_limit=2):
    """
    Recursively explore and print the structure of a JSON object
    """
    if depth >= max_depth:
        print(f"{prefix}... (max depth reached)")
        return
        
    if isinstance(data, dict):
        print(f"{prefix}Dict with {len(data)} keys")
        
        # Get a sample of keys
        keys = list(data.keys())
        sample_keys = keys[:min(sample_limit, len(keys))]
        
        print(f"{prefix}Sample keys: {sample_keys}")
        
        # Explore a few items
        for i, key in enumerate(sample_keys):
            print(f"{prefix}Key: {key}")
            explore_structure(data[key], prefix=prefix+"  ", depth=depth+1, 
                             max_depth=max_depth, sample_limit=sample_limit)
            
    elif isinstance(data, list):
        print(f"{prefix}List with {len(data)} items")
        
        # Explore a few items
        for i in range(min(sample_limit, len(data))):
            print(f"{prefix}Item {i}:")
            explore_structure(data[i], prefix=prefix+"  ", depth=depth+1, 
                             max_depth=max_depth, sample_limit=sample_limit)
            
    else:
        # For primitive values
        value_str = str(data)
        if len(value_str) > 50:
            value_str = value_str[:50] + "..."
        print(f"{prefix}Value: {type(data).__name__} = {value_str}")

```
